```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
COSMOS = pd.read_csv('FINAL VERSION_match_COSMOS_del_rows_99(1)-Copy1.csv')
COSMOS_1=COSMOS[COSMOS>0]
```

```python
def HISTO(x,df=COSMOS_1):
    fig = plt.figure(figsize=(10,8))
    plt.hist(df[x],bins=200,histtype='bar',facecolor='blue',lw=4)
    plt.xlabel('value')
    plt.ylabel('counts')
    plt.title(x,fontsize=25)
    plt.savefig()
    return fig
```
